```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Salary_Data.csv")
data.head()
```

Transpose of a matrix multiplied by itself is a Square Matrix: $X^{T}X = Square Matrix$

Inverse of a matrix multiplied by itself is an Identity Matrix: $(X^{T}X)^{-1} * X^{T}X  = Identity Matrix$

Solve for B which is a matrix of Coefficients

$y = XB + e$

$y = XB (set e to zero to minimize it)$

$X^{T}y = X^{T}XB$

$((X^{T}X)^{-1})X^{T}y = (X^{T}X)^{-1}(X^{T}X)B $

$B = (X^{T}X)^{-1}X^{T}y$

```python
X = data[["YearsExperience"]].copy()
Y = data[["Salary"]].copy()
```

```python
X.head()
```

```python
Y.head()
```

```python
X["Intercept"] = 1
X = X[["Intercept", "YearsExperience"]]
```

```python
X
```

```python
X_T = X.T
X_T
```

```python
B = np.linalg.inv(X_T @ X) @ X_T @ Y
B
```

```python
B.index = X.columns
B
```

```python
predictions = X @ B
```

```python
predictions
```

$SSR = \sum (y- \hat{y})^2$

$SST = \sum (y- \bar{y})^2$

$r^2 = 1 - (SSR / SST)$

```python
SSR = ((Y - predictions) ** 2).sum()
```

```python
SSR
```

```python
SST = ((Y - Y.mean()) ** 2).sum()
```

```python
SST
```

```python
R2 = 1 - (SSR/SST)
```

```python
R2
```

```python
print(f"SSR: {SSR}")

print(f"SST: {SST}")

print(f"R2: {R2}")

print(B)
```
